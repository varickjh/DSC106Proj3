```python
dexcom_df = pd.read_csv('Dexcom_001.csv')
dexcom_df = dexcom_df[12:].drop(columns=['Index'])
dexcom_df['ID'] = int('001')

dexcom_df

ids = ['002', '003', '004', '005', '006', '007', '008', '009', '010', '011', '012', '013', '014', '015', '016']
for id in ids:
    temp_df = pd.read_csv(f"Dexcom_{id}.csv")
    temp_df = temp_df[12:].drop(columns=['Index'])
    temp_df['ID'] = int(id)
    dexcom_df = pd.concat([dexcom_df, temp_df], ignore_index=True)

dexcom_df = dexcom_df.rename(columns={'Timestamp (YYYY-MM-DDThh:mm:ss)': 'Timestamp'})
dexcom_df = dexcom_df.rename(columns={'Glucose Value (mg/dL)': 'Glucose'})
dexcom_df['Timestamp'] = pd.to_datetime(dexcom_df['Timestamp'], errors='coerce')
dexcom_df['Glucose'] = pd.to_numeric(dexcom_df['Glucose'], errors='coerce')

dexcom_df
```

```python
food_df = pd.read_csv('Food_Log_001.csv')
food_df['ID'] = int('011')

ids = ['002', '003', '004', '005', '006', '007', '008', '009', '010', '011', '012', '013', '014', '015', '016']
for id in ids:
    temp_df = pd.read_csv(f"Food_Log_{id}.csv")
    temp_df['ID'] = int(id)
    food_df = pd.concat([food_df, temp_df], ignore_index=True)

food_df = food_df.drop(columns =['2020-02-22', '10:30:00',
       '2020-02-22 10:30:00', 'Chicken Nuggets', '8.0', 'piece',
       'Chicken Nuggets.1', '393.0', '19.0', '0.1', '20.0'], axis=1)
food_df
```

```python
# Convert time_begin to datetime
food_df['time_begin'] = pd.to_datetime(food_df['time_begin'], errors='coerce')

# FDA Daily Values (based on 2000 kcal diet)
daily_values = {
    'total_carb': 275,    # grams
    'protein': 50,        # grams
    'sugar': 50,          # grams
    'total_fat': 78,       # grams
    'dietary_fiber': 28   # grams
}
thresholds = {nutrient: 0.05 * dv for nutrient, dv in daily_values.items()}

# Add boolean flag columns
for nutrient, threshold in thresholds.items():
    food_df[f'{nutrient}_present'] = food_df[nutrient] >= threshold

# Preview result
#print(food_df[['ID', 'time_begin', 'logged_food', 'total_carb', 'protein', 'sugar', 'total_fat',
#               'total_carb_present', 'protein_present', 'sugar_present', 'total_fat_present']].head())
food_df
```

```python
# Function to filter based on selection
def get_avg_glucose_peaks(food_df, dexcom_df, selected_features, window_hrs=2):
    feature_map = {
        1: 'total_carb_present',
        2: 'protein_present',
        3: 'sugar_present',
        4: 'total_fat_present',
        5: 'dietary_fiber_present'
    }

    selected_flags = [feature_map[f] for f in selected_features if f in feature_map]
    all_flags = [col for col in food_df.columns if col.endswith('_present')]
    excluded_flags = list(set(all_flags) - set(selected_flags))

    condition_include = food_df[selected_flags].all(axis=1)
    condition_exclude = ~food_df[excluded_flags].any(axis=1)
    filtered_meals = food_df[condition_include & condition_exclude].copy()

    results = []
    for pid in filtered_meals['ID'].unique():
        meals = filtered_meals[filtered_meals['ID'] == pid]
        glucose_data = dexcom_df[dexcom_df['ID'] == pid]

        peaks = []
        for _, meal in meals.iterrows():
            start = meal['time_begin']
            end = start + pd.Timedelta(hours=2)
            window = glucose_data[(glucose_data['Timestamp'] >= start) & (glucose_data['Timestamp'] <= end)]
            if not window.empty:
                peaks.append(window['Glucose'].max())

        if peaks:
            results.append({
                'ID': pid,
                'Avg_Glucose_Peak': sum(peaks) / len(peaks),
                'Meal_Count': len(peaks)
            })

    return pd.DataFrame(results)

peaks_df = get_avg_glucose_peaks(food_df, dexcom_df, selected_features=[1,2,3,4,5])  #testing
print(peaks_df.to_string(index=False))
```
